```python
# Dependencies 
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#  set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# visit the mars website 
url = 'https://redplanetscience.com/'
browser.visit(url)

# convert the browser html to a soup object 
html = browser.html
news_soup = soup(html, 'html.parser')
```

```python
print(news_soup.prettify())
```

```python
slide_elem = news_soup.select_one('div.list_text')
```

```python
# display the current title content
slide_elem.find('div', class_='content_title')
```

```python
# finding the first tag and save it as 'new_title'
news_title = slide_elem.find('div', class_='content_title').get_text()
news_title
```

```python
# finding the paragraph text
news_p = slide_elem.find('div', class_='article_teaser_body').get_text()
news_p
```

```python
# visit image URl 
url = 'https://spaceimages-mars.com/'
browser.visit(url)
```

```python
# find and click the full image button
full_image_link = browser.find_by_tag('button')
# go to the full image url
full_image_link[1].click()
```

```python
# visit hemisphere URl 
hemis_url = 'https://marshemispheres.com/'
browser.visit(hemis_url)
```

```python
# create a list of the images and titles
hemis_img_url = []

# get a list of all of hemispeheres
links = browser.find_by_css('a.product-item img')

# looping thru the links, finding the sample anchor, reurning the href
for i in range(len(links)):
    
    # hemisphere dictionary
    hemisphere = {}
    
    # finding the element on each loop
    browser.find_by_css('a.product-item img')[i].click()
    
    # finding sample image  and extract href
    sample_element = browser.find_by_text("Sample").first
    hemisphere["img_url"] = sample_element["href"]
    
    # getting Hemisphere title
    hemisphere["title"] = browser.find_by_css("h2.title").text
    
    # append hemisphere object to list
    hemis_img_url.append(hemisphere)

    # navigate back
    browser.back()
```
